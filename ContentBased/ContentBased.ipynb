```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits = pd.read_csv('dataset/tmdb_5000_credits.csv')
movies_df = pd.read_csv('dataset/tmdb_5000_movies.csv')
print("Credits:",credits.shape)
print("Movies Dataframe:",movies_df.shape)
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
credits = credits.rename(index=str, columns={"movie_id": "id"})
movies = movies_df.merge(credits, on='id')
movies.head()
```

```python
movies.columns
```

```python
movies.drop(['production_companies','title_x','title_y','status'],axis=1,inplace=True)
movies.info()
```

```python
import ast
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    movies[feature] = movies[feature].map(lambda x: ast.literal_eval(x))

def make_genresList(x):
    if isinstance(x, list):
        for i in x:
            if i['name'] == 'Science Fiction':
                i['name'] = 'Sci-Fi'
        names = [i['name'] for i in x]
        return names
    return []
movies['genres'] = movies['genres'].map(lambda x: make_genresList(x))

def make_directorList(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
movies['director'] = movies['crew'].map(lambda x: make_directorList(x))

def make_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []

features = ['cast', 'keywords']
for feature in features:
    movies[feature] = movies[feature].map(lambda x: make_list(x))
```

### Overview(Plot) Based Recommendation System

```python
movies1 = movies.copy()
```

```python
movies1['overview'].head(2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tvf = TfidfVectorizer(min_df=3, max_features=None,strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',
                      ngram_range=(1,3),stop_words= 'english')
movies1['overview'] = movies1['overview'].fillna('')
```

```python
tvf_matrix = tvf.fit_transform(movies1['overview'])
```

```python
tvf_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tvf_matrix,tvf_matrix)
```

```python
sig[0]
```

```python
indices = pd.Series(movies1.index, index = movies1['original_title']).drop_duplicates()
```

```python
indices.head(15)
```

```python
def recommendation1(title,sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x:x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    data =  {
                'title':[movies1['original_title'].iloc[i] for i in movie_indices],
                'score':[sig_scores[j][1] for j in range(0,len(sig_scores))],
                'genres':[movies1['genres'].iloc[i] for i in movie_indices],
                'keywords':[movies1['keywords'].iloc[i] for i in movie_indices]
            }
    df = pd.DataFrame(data)
    return df
```

```python
recommendation1('Avatar')
```

```python
recommendation1('Avengers: Age of Ultron')
```

### Genres,Keywords,Cast and Crew Based Recommendation System

```python
movies[['original_title', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
movies2 = movies.copy()
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
        
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    movies2[feature] = movies2[feature].apply(clean_data)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
movies2['soup'] = movies2.apply(create_soup, axis=1)
```

```python
movies2['soup'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(movies2['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
movies2 = movies2.reset_index()
indices = pd.Series(movies2.index, index=movies2['original_title'])
```

```python
movies2.head(2)
```

```python
def recommendation2(title,cosine_sim=cosine_sim):
    idx = indices[title]
    cos_scores = list(enumerate(cosine_sim[idx]))
    cos_scores = sorted(cos_scores, key=lambda x:x[1], reverse=True)
    cos_scores = cos_scores[1:11]
    movie_indices = [i[0] for i in cos_scores]
    data =  {
                'title':[movies['original_title'].iloc[i] for i in movie_indices],
                'score':[cos_scores[j][1] for j in range(0,len(cos_scores))],
                'genres':[movies['genres'].iloc[i] for i in movie_indices],
                'keywords':[movies['keywords'].iloc[i] for i in movie_indices],
                'cast':[movies['cast'].iloc[i] for i in movie_indices],
                'director':[movies['director'].iloc[i] for i in movie_indices]
            }
    df = pd.DataFrame(data)
    return df
```

```python
recommendation2('Avengers: Age of Ultron')
```

```python
recommendation2('Superman Returns')
```

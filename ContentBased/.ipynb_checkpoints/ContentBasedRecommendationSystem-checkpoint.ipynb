```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')
print("Credits:",credits.shape)
print("Movies Dataframe:",movies_df.shape)
```

```python
movies_df.head()
```

```python
credits = credits.rename(index=str, columns={"movie_id": "id"})
movies = movies_df.merge(credits, on='id')
movies.head()
```

```python
movies.columns
```

```python
movies.drop(['production_companies','title_x','title_y','status'],axis=1,inplace=True)
movies.info()
```

```python
import ast
movies['genres'] = movies['genres'].map(lambda x: ast.literal_eval(x))

def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tvf = TfidfVectorizer(min_df=3, max_features=None,strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',
                      ngram_range=(1,3),stop_words= 'english')
movies['overview'] = movies['overview'].fillna('')
```

```python
tvf_matrix = tvf.fit_transform(movies['overview'])
```

```python
tvf_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tvf_matrix,tvf_matrix)
```

```python
sig[0]
```

```python
indices = pd.Series(movies.index, index = movies['original_title']).drop_duplicates()
```

```python
indices.head(15)
```

```python
def recommendation(title,sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x:x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    data =  {
                'title':[movies['original_title'].iloc[i] for i in movie_indices],
                'score':[sig_scores[j][1] for j in range(0,len(sig_scores))]
            }
    df = pd.DataFrame(data)
    return df
```

```python
recommendation('Avatar')
```

```python
recommendation('Avengers: Age of Ultron')
```
